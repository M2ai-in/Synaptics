# Planner Tool


```python
from agentpro.tools.base import Tool
from agentpro import AgentPro, ares_tool, code_tool, youtube_tool, slide_tool

sub_agent = AgentPro(tools=[ares_tool, code_tool, youtube_tool, slide_tool])

class PlannerTool(Tool):
    name: str = "teacher_agent_tool"
    description: str = (
        "Acts as a teacher assistant: recommends YouTube videos, summarizes content, "
        "plans learning paths, and helps make study notes for any topic."
    )
    arg: str = "Instruction like 'plan deep learning', 'recommend videos on transformers', or 'summarize a topic'."

    def run(self, prompt: str) -> str:
        intent, _, topic = prompt.partition(" ")
        intent = intent.strip().lower()
        topic = topic.strip()

        if not topic:
            return "Please provide a topic along with the instruction."

        if intent in ["plan", "learning", "learn"]:
            return self.plan_learning(topic)
        elif intent in ["recommend", "video", "videos"]:
            return self.recommend_videos(topic)
        elif intent in ["summarize", "notes", "note"]:
            return self.summarize_and_note(topic)
        else:
            return "I didn't understand that. Try: 'plan deep learning', 'recommend videos on transformers', or 'summarize quantum mechanics'."

    def plan_learning(self, topic: str) -> str:
        # Use `ares_tool` via sub_agent to decompose topic
        prompt = f"Break down the topic '{topic}' into a step-by-step learning plan with subtopics."
        return sub_agent(prompt)

    def recommend_videos(self, topic: str) -> str:
        # Use `youtube_tool`
        prompt = f"Find top YouTube videos to learn about {topic} effectively."
        return sub_agent(prompt)

    def summarize_and_note(self, topic: str) -> str:
        # Use `ares_tool` + `slide_tool` to fetch content and make notes
        prompt = f"Search and summarize key points from online resources and make concise notes on {topic}."
        return sub_agent(prompt)
```

```python
from agentpro.tools.base import Tool
from typing import List, Any

class NoteSearchTool(Tool):
    name: str = "note_search_tool"
    description: str = (
        "Searches for relevant study notes or summaries from a vector database "
        "based on semantic similarity to your query."
    )
    arg: str = "A topic or question to search for, like 'backpropagation' or 'uses of dropout in deep learning'."

    def __init__(self, vector_db: Any, embedding_model: Any):
        super().__init__()
        object.__setattr__(self, 'vector_db', vector_db)
        object.__setattr__(self, 'embedding_model', embedding_model)

    def run(self, prompt: str) -> str:
        query = prompt.strip()
        if not query:
            return "Please provide a topic or query to search notes."

        query_embedding = self.embed(query)

        results = self.vector_db.similarity_search(query_embedding, k=5)

        return self.format_results(results)

    def embed(self, text: str) -> List[float]:
        return self.embedding_model.encode(text, normalize_embeddings=True).tolist()

    def format_results(self, results: List[dict]) -> str:
        if not results:
            return "No matching notes found."

        output = ["🗒️ Top Matching Notes:\n"]
        for i, item in enumerate(results, 1):
            content = item.get("text", "No content.")
            score = item.get("score", "N/A")
            output.append(f"{i}. {content.strip()}  \n_(score: {score:.4f})_\n")
        return "\n".join(output)
```

```python
import os
import requests
from youtube_transcript_api import YouTubeTranscriptApi
from pdfminer.high_level import extract_text
from bs4 import BeautifulSoup
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

class AutoNoteUpdater:
    def __init__(self, embedding_model, faiss_index, note_store):
        self.embedding_model = embedding_model
        self.index = faiss_index
        self.note_store = note_store

    def embed_and_add(self, texts: list):
        embeddings = self.embedding_model.encode(texts, normalize_embeddings=True)
        embeddings = np.array(embeddings).astype("float32")
        self.index.add(embeddings)
        for text in texts:
            self.note_store.append({"text": text})

    def from_youtube(self, video_id: str):
        try:
            transcript = YouTubeTranscriptApi.get_transcript(video_id)
            full_text = " ".join([x["text"] for x in transcript])
            self.embed_and_add([full_text])
        except Exception as e:
            print(f"Failed to fetch YouTube transcript: {e}")

    def from_pdf(self, file_path: str):
        try:
            text = extract_text(file_path)
            self.embed_and_add([text])
        except Exception as e:
            print(f"Failed to extract PDF: {e}")

    def from_url(self, url: str):
        try:
            html = requests.get(url).text
            soup = BeautifulSoup(html, "html.parser")
            paragraphs = soup.find_all("p")
            full_text = " ".join([p.get_text() for p in paragraphs])
            self.embed_and_add([full_text])
        except Exception as e:
            print(f"Failed to fetch URL: {e}")

    def from_slide_texts(self, slide_texts: list):
        self.embed_and_add(slide_texts)
```

```python
from agentpro.tools import YouTubeSearchTool
from agentpro.tools.base import Tool
from typing import List, Any
import numpy as np

class NoteManager(Tool):
    name: str = "note_manager"
    description: str = (
        "Searches, embeds, and retrieves notes from various sources (manual, YouTube, etc.)"
    )
    arg: str = "A query or topic like 'what is dropout in neural nets?'"

    def __init__(self, vector_db: Any, embedding_model: Any, youtube_tool: Any):
        super().__init__()
        object.__setattr__(self, "vector_db", vector_db)
        object.__setattr__(self, "embedding_model", embedding_model)
        object.__setattr__(self, "youtube_tool", youtube_tool)

    def run(self, prompt: str) -> str:
        if not prompt.strip():
            return "Please enter a valid query."

        query_embedding = self.embed(prompt)
        search_results = self.vector_db.similarity_search(query_embedding, k=5)

        if not search_results:
            return "No notes found. Try uploading or extracting from YouTube."

        return self.format_results(search_results)

    def embed(self, text: str) -> List[float]:
        return self.embedding_model.encode(text, normalize_embeddings=True).tolist()

    def format_results(self, results: List[dict]) -> str:
        output = ["🗒️ Top Matching Notes:\n"]
        for i, item in enumerate(results, 1):
            content = item.get("text", "No content.")
            score = item.get("score", "N/A")
            output.append(f"{i}. {content.strip()}  \n_(score: {score:.4f})_\n")
        return "\n".join(output)

    def ingest_youtube_notes(self, query: str):
        try:
            print(f"Fetching YouTube content for query: {query}")
            result = self.youtube_tool.run(query)
            if isinstance(result, str):
                self.vector_db.note_store.append({"text": result})
                video_embedding = self.embedding_model.encode([result], normalize_embeddings=True)
                self.vector_db.index.add(np.array(video_embedding).astype("float32"))
                return "YouTube content ingested into notes."
            return "Unexpected response format from YouTube tool."
        except Exception as e:
            return f"Error during YouTube ingestion: {e}"
```

```python
from agentpro import ares_tool, youtube_tool
from agentpro.tools.base import Tool
from sentence_transformers import SentenceTransformer
from typing import List, Any
import numpy as np
import faiss

# FAISS-based vector database
class FAISSVectorDB:
    def __init__(self, index, note_store):
        self.index = index
        self.note_store = note_store

    def similarity_search(self, query_embedding, k=5):
        query_vector = np.array(query_embedding).astype("float32").reshape(1, -1)
        scores, indices = self.index.search(query_vector, k)
        results = []
        for i, idx in enumerate(indices[0]):
            if idx < len(self.note_store):
                results.append({
                    "text": self.note_store[idx]["text"],
                    "score": float(scores[0][i])
                })
        return results

# NoteManager with YouTube ingestion and summarization
class NoteManager(Tool):
    name: str = "note_manager"
    description: str = (
        "Searches and summarizes top relevant note from YouTube or stored data."
    )
    arg: str = "A question or topic like 'explain transformers'"

    def __init__(self, vector_db: Any, embedding_model: Any, youtube_tool: Any):
        super().__init__()
        object.__setattr__(self, "vector_db", vector_db)
        object.__setattr__(self, "embedding_model", embedding_model)
        object.__setattr__(self, "youtube_tool", youtube_tool)

    def run(self, prompt: str) -> str:
        if not prompt.strip():
            return "Please enter a valid query."

        query_embedding = self.embed(prompt)
        results = self.vector_db.similarity_search(query_embedding, k=1)

        if not results:
            return "No matching notes found. Try ingesting from YouTube first."

        top_note = results[0]["text"]
        try:
            summary = ares_tool.run(top_note)
            return f"**Summary of top note:**\n{summary}"
        except Exception as e:
            return f"Found a note but summarization failed: {e}"

    def embed(self, text: str) -> List[float]:
        return self.embedding_model.encode(text, normalize_embeddings=True).tolist()

    def ingest_youtube_notes(self, query: str) -> str:
        try:
            print(f"🔎 Fetching YouTube content for query: {query}")
            result = self.youtube_tool.run(query)
            if isinstance(result, str) and result.strip():
                self.vector_db.note_store.append({"text": result})
                video_embedding = self.embedding_model.encode([result], normalize_embeddings=True)
                self.vector_db.index.add(np.array(video_embedding).astype("float32"))
                return " YouTube content ingested successfully."
            return " Unexpected YouTube result format."
        except Exception as e:
            return f" Error during ingestion: {e}"
```

```python

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
note_store = []
faiss_index = faiss.IndexFlatIP(embedding_model.get_sentence_embedding_dimension())
vector_db = FAISSVectorDB(faiss_index, note_store)


note_manager = NoteManager(vector_db=vector_db, embedding_model=embedding_model, youtube_tool=youtube_tool)

# Step 1: Ingest a YouTube video
print(note_manager.ingest_youtube_notes("transformer architecture explained"))

# Step 2: Ask a question from the notes
response = note_manager.run("How does attention work in transformers?")
print(response)
```
